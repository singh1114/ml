```python
import numpy as np
```

```python
import pandas as pd
```

### Creating DF using dictionary

```python
d = {'A': [1, 2, np.nan], 'B': [3, np.nan, np.nan], 'C': [4, 5, 6]}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

### Droping null values

```python
df.dropna()
```

```python
# Dropping null valued columns
df.dropna(axis=1)
```

```python
# Dropping with threshold
df.dropna(thresh=2)
```

### Filling null values

```python
df.fillna(value='String')
```

```python
df['A'].fillna(value=df['A'].mean())
```
